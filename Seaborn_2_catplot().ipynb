## Categoricalplot()
- categorical scatter plot:
    1. stripplot
    2. swarmplot
- categorical distribution plot:
    1. Boxplot
    2. violinplot
- categorical estimate plot:
    1. Barplot
    2. Countplot

- categorical estimate plot:
    - Countplot: Used for univariate analysis and it is used for categorical data. It is same as 
        value_counts()-tells frequency of unique elements in the data.
        - for numeric data we use histogram, univariate only
        - barplot: categorical to numerical, bivariate analysis
        - scatterplot: numeric to numeric, bivariate analysis
- categorical - categorical : use countplot() with hue

```python
df=sns.load_dataset("tips")
df.head(2)
```

```python
# how many males and females visited the restaurant
# in pandas we will use value_counts(), now use countplot()
sns.countplot(x="sex",data=df);
```

```python
sns.countplot(x="smoker",data=df);
```

```python
# at what time customer visited the restaurant most(peak time)
sns.countplot(x="time",data=df);
```

```python
#changing the order
sns.countplot(x="time",data=df,order=["Dinner","Lunch"]);
```

```python
# find out the peak day
sns.countplot(x="day",data=df);
## peak day was saturday and off day was friday
```

```python
## on which day males and females visited more
sns.countplot(x="day",hue="sex",data=df); ##doing bivariate analysis
# females visited more on thursday and males visited more on sat and sun
```

```python
## how many smoker and non-smoker in males and females
sns.countplot(x="smoker",hue="sex",data=df); 
## male and female category both are more non-smoker than smoker
```

# Barplot() :
- used for bivariate analysis when 1 col is categorical and other is numerical
- to compare the categories
- matplotlib barplot- gives max
- sns barplot- gives average
- it will give all

```python
## find out who(sex-categorical) paid maximum bill(numeric)
# cat-num: use barplot
## on x-axis-category always
sns.catplot(x="sex",y="total_bill",data=df,kind="bar"); ## by default estimator is based on average
## black line is confidence interval , part of stats
```

```python
df.groupby(["sex"])["total_bill"].mean()
```

```python
sns.catplot(x="sex",y="total_bill",data=df,kind="bar",estimator=sum); ## sum
```

```python
sns.catplot(x="sex",y="total_bill",data=df,kind="bar",estimator=max);
## max bill is 50 using estimator
```

## catplot is general plot for categorical

```python
sns.barplot(x="sex",y="total_bill",data=df,estimator=max);
```

```python
## on which day tip given by customer was max
sns.barplot(x="day",y="tip",data=df,estimator=max);
## on saturday max tip was given
```

```python
## who paid the highest tip based on average on each day
sns.catplot(x="day",y="tip",data=df,hue="sex",kind="bar");
## avg tip was given by male and female on thurs was around 3 and 2.5 
```

```python
## check when smoker and non smoker came most 
sns.countplot(x="time",hue="smoker",data=df);
```

```python
## find out who paid the max bill, whether a smoker and non smoker
sns.barplot(x="smoker",y="total_bill",data=df);
#smoker paid the max bill
```

### Correlation: tells the relation between two variables, value of correlation lies between -1 to 1.
- If correlation value is 0, that means both are independent, no relation between them. (like location and salary)
- If value is > 0, means positive correlation(if 1 increases, other increases and if 1 decreases then other have to decrease too) 
- if value < 0, means negative correlation 
- close to 1, highly positively correlated
- close to -1, highly negatively correlated


```python
df.corr() ## correlation in matrix form 
## only for numeric columns
## moderately(sort of) linear relation b/w tip and total_bill
## no negative correlation 
```
