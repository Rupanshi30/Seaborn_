Seaborn is used for statistical analysis

It allows us to do multivariate analysis(more than two columns)

## 1. Numerical Plots()
1. relplot()
2. scatterplot()
3. lineplot()

### Scatter Plot()- for bivariate analysis and both columns are numeric(continuous type)
 To plot scatterplot, we will use
- relplot()
- scatterplot()

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
df.dtypes
```

```python
## fetch the trend of total bill and tip
sns.relplot(x="total_bill",y="tip",data=df)
#insights : it is showing sort of linear relation, if bill increases tip increases too
# average tip was between 2-4
```

```python
sns.relplot(x="total_bill",y="tip",hue="sex",data=df);  ##multivariate analysis
# insights : max bill and tip was paid by a male 
            # min bill and tip was paid by a female 
            # max bill paid by a female was around 45 but tip was around 2.3 only
        
```

## hue, style and size will take only categorical type(sex,smoker,day,time)

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=df);
# insights : max bill payer was a smoker 
```

```python
sns.relplot(x="total_bill",y="tip",hue="sex",data=df,style="smoker");# style only takes categorical data
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",data=df);
# less family size, less bill and less tip
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",data=df);
# people visited more in dinner time and gave more tip
# people who visited on lunch time gave avg tip of 2
```

```python
sns.relplot(x="total_bill",y="tip",hue="sex",data=df,style="smoker",size="size",sizes=(50,200)); 
# size only takes categorical data
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df);
```

```python
## Difference between relplot and scatterplot
# relplot is generic(can make lineplot and scatterplot), by default it gave scatterplot
```

```python
## two ways to create lineplot :
    #relplot means relational plot and lineplot
```

```python
sns.relplot(x="total_bill",y="tip",data=df,kind="line")
```

```python
sns.lineplot(x="total_bill",y="tip",data=df);
```

```python
df1=sns.load_dataset("flights")
df1
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="sex",col="day");
## friday was not the peak day
# female visited more on thursday
# max bill payer came on saturday
# people visited more on sat, sun
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="sex",col="day",col_wrap=2); ##col_wrap=2 means 2 plot per columns
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="sex",col="time");
## people visited more on dinner time 
```
